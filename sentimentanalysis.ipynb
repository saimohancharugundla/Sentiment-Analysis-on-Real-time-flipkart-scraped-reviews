```python
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
```

```python
import emoji
import re
```

```python
def give_emoji_free_text(comm):
    allchars = [str for str in comm]
    emoji_list = [c for c in allchars if c in emoji.UNICODE_EMOJI]
    clean_text = ' '.join([str for str in comm.split() if not any(i in str for i in emoji_list)])
    return clean_text
```

```python
comm = [give_emoji_free_text(txt) for txt in comm]
comm = [re.sub("READ MORE","",i) for i in comm]
comm = [re.sub('-','',i) for i in comm]
comm = [re.sub('\d','',i) for i in comm]
comm = [re.sub('GB|gb','',i) for i in comm]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stopw = (stopwords.words('english'))
setstop = ['money','purchase','penny','market','product','every']
stopw.extend(setstop)
stopw = set(stopw)
stopw
```

```python
from textblob import TextBlob
blob1 = TextBlob(comm[9])
blob1.sentiment
```

```python
v = []
for val in text:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    v.append(" ".join(tokens))
for word in range(len(v)):
    token = word_tokenize(v[word])
    for tkn in token:
        if tkn in stopw:
            token.remove(tkn)
    v[word] = " ".join([str(ele) for ele in token])
    
b = []
for val in comm:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    b.append(" ".join(tokens))
for word in range(len(b)):
    token = word_tokenize(b[word])
    for tkn in token:
        if tkn in stopw:
            token.remove(tkn)
    b[word] = " ".join([str(ele) for ele in token])
```

```python
from textblob import TextBlob
score = []
for i in range(len(v)):
    bl = TextBlob(v[i])
    score.append(bl.sentiment.polarity)
    
from textblob import TextBlob
cscore = []
for i in range(len(b)):
    bl = TextBlob(b[i])
    cscore.append(bl.sentiment.polarity)

```

```python
for i in range(len(score)):
    if score[i] < 0.0:
        score[i] = 0
    elif score[i] > 0.0:
        score[i] = 1
    elif score[i] == 0.0:
        score[i] = 0
        
for i in range(len(cscore)):
    if cscore[i] < 0.0:
        cscore[i] = 0
    elif cscore[i] > 0.0:
        cscore[i] = 1
    elif cscore[i] == 0.0:
        cscore[i] = 0
```

```python
data['Score'] = score
data['Cscore'] = cscore
```
